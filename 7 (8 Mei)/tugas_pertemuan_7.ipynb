```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

# Baca data dari file CSV
data = pd.read_csv('dataset_3 (tgl 8 mei).csv')
print("Dataset shape:", data.shape)
print("Columns:", data.columns.tolist())
print("\nTarget distribution:")
print(data['Jenis Kendaraan Dipilih'].value_counts())

# Encode categorical features
le_harga = LabelEncoder()
le_status = LabelEncoder()
le_kapasitas = LabelEncoder()

# Pilih fitur yang akan digunakan
data['Preferensi_Harga_Encoded'] = le_harga.fit_transform(data['Preferensi Harga'])
data['Status_Pekerjaan_Encoded'] = le_status.fit_transform(data['Status Pekerjaan'])
data['Preferensi_Kapasitas_Encoded'] = le_kapasitas.fit_transform(data['Preferensi Kapasitas'])

# Pisahkan fitur (x) dan target (y) dari data
x = data[['Preferensi_Harga_Encoded', 'Status_Pekerjaan_Encoded', 'Preferensi_Kapasitas_Encoded']].values
y = data['Jenis Kendaraan Dipilih'].values

# Train model
clf = MultinomialNB()
clf = clf.fit(x, y)

# Evaluasi akurasi
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x)
accuracy = accuracy_score(y, y_pred)
print(f"\nAkurasi model: {accuracy:.2%}")

def prog():
    print("\n=== PREDIKSI JENIS KENDARAAN ===")
    print("Preferensi Harga: 0=Murah, 1=Sedang, 2=Mahal")
    print("Status Pekerjaan: 0=Tidak Bekerja, 1=Pegawai Negeri, 2=Pelajar/Mahasiswa, 3=Pensiunan, 4=Wiraswasta")
    print("Preferensi Kapasitas: 0=Besar, 1=Kecil, 2=Sedang")
    
    harga = int(input("Preferensi Harga (0-2): "))
    status = int(input("Status Pekerjaan (0-4): "))
    kapasitas = int(input("Preferensi Kapasitas (0-2): "))
    
    # Prediksi
    prediksi = clf.predict([[harga, status, kapasitas]])
    probabilitas = clf.predict_proba([[harga, status, kapasitas]])
    
    print(f"\nPrediksi: {prediksi[0]}")
    print("Probabilitas untuk setiap kendaraan:")
    for i, kendaraan in enumerate(clf.classes_):
        print(f"- {kendaraan}: {probabilitas[0][i]:.3f}")

prog()
```
